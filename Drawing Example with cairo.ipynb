To get better looking images in IPython notebook you need [cairo](http://cairographics.org) installed.

For windows you can get pycairo [here](http://www.lfd.uci.edu/~gohlke/pythonlibs/#pycairo)

[Dependencies](http://cairographics.org/download/):
- libcairo-2.dll
- libpng13.dll
- zlib1.dll

And the must be on the PATH.

Note: In my case all of these dlls were already present due to some other apps so no need to install them.
If you dont have them you need to follow [instructions](http://www.gtk.org/download/windows.php) and download [MSYS2](https://msys2.github.io/) which contains these dependencies.
